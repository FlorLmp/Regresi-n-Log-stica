```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
import scipy.optimize as curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#Cargamos el archivo csv
dm = pd.read_csv("Florencia_limpio_para_regresion.csv")
dm = dm.drop([ 'Unnamed: 0',], axis=1)
dm.head(5)
```

Apartado de categorización

*host_response_time*

```python
#Categorizamos la variable "host_response_time", donde 1 = "rápido" y 0 = "lento"
dm["host_response_time"] = dm["host_response_time"].replace({
'within an hour': 1,
'within a few hours': 1,
'within a day': 1,
'a few days or more': 0,
'Sin tiempo': 0,
'Desconocido': 0
})
```

*host_response_rate*

```python
dm['host_response_rate'] = dm['host_response_rate'].str.replace('%', '', regex=True).astype(float) / 100
```

*host_acceptance_rate*

```python
dm['host_acceptance_rate'] = dm['host_acceptance_rate'].str.replace('%', '', regex=True).astype(float) / 100
```

*host_is_superhost*

```python
dm['host_is_superhost'] = dm['host_is_superhost'].replace({'f': 0, 't': 1})
```

*host_identity_verified*

```python
#Convertimos Los booleanos a números
dm['host_identity_verified'] = dm['host_identity_verified'].replace({'f': 0, 't': 1})
```

*room_type*

```python
#categorizamos la variable "room type", donde 1 = "privado" y 0 = "compartido"
dm["room_type"] = dm["room_type"].replace({
'Entire home/apt': 1,
'Private room': 0,
'Shared room': 0,
'Hotel room': 0
})
```

*has_availability*

```python
#Convertimos Los booleanos a números
dm['has_availability'] = dm['has_availability'].replace({'f': 0, 't': 1})
```

*instant_bookable*

```python
#Convertimos Los booleanos a números
dm['instant_bookable'] = dm['instant_bookable'].replace({'f': 0, 't': 1})
```

*accommodates*

```python
#categorizamos la variable "room type", donde 1 = "privado" y 0 = "compartido"
dm["accommodates"] = dm["accommodates"].replace({
    1: 0,
    2: 0,
    3: 1,
    3.6: 1,
    4: 1,
    5: 1,
    6: 1,
    7: 1
})
```

*price*

```python
valor_euro=18.50
dm["price_mxn"]=dm["price"] * valor_euro
```

```python
valor_euro = 18.50
dm["price_mxn"] = dm["price"] * valor_euro

#Categorizamos la variable "price" en 0="barato" y 1="caro"
median_price = dm["price_mxn"].median()
dm["precio"] = (dm["price_mxn"] > median_price).astype(int)
```

```python
dm["price_mxn"] = dm["price_mxn"].apply(lambda x: 1 if x > 1093 else 0)
```

*review_scores_rating*

```python
median_rating = dm["review_scores_rating"].median()
dm["review_rating"] = (dm["review_scores_rating"] > median_rating).astype(int)
```

```python
dm["review_scores_rating"] = dm["review_scores_rating"].apply(lambda x: 1 if x > 4.8 else 0)
```
